```python
X = dataset.iloc[:,:64].values
y = dataset.iloc[:,64:66].values
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(42, input_dim=64, activation='relu'))
model.add(Dense(24, activation='relu'))
model.add(Dense(2, activation='linear'))
```

```python
#model.compile(loss='mean_squared_error', optimizer='sgd', metrics=['accuracy']) #regresion
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) #categorical
```

```python
history = model.fit(X_train, y_train, epochs=10, batch_size=64)
```

```python
y_pred = model.predict(X_test)
```
