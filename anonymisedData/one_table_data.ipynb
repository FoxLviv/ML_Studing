```python
import pandas as pd
```

```python
student_info= pd.read_csv("studentInfo.csv")
courses = pd.read_csv("courses.csv")
studentAssessment=pd.read_csv("studentAssessment.csv")
assessments =pd.read_csv("assessments.csv")
studentRegistration = pd.read_csv("studentRegistration.csv")
student_Vle = pd.read_csv("studentVle.csv")
vle = pd.read_csv("vle.csv")
```

```python
C_A=pd.merge(courses, assessments)
C_A_Sa = pd.merge(C_A, studentAssessment)
C_A_Sa_Si = pd.merge(C_A_Sa, student_info)
```

```python
C_A_Sa_Si_V = pd.merge(C_A_Sa_Si, vle)
C_A_Sa_Si_V
```

```python
C_A_Sa_Si_V_Sv = pd.merge(C_A_Sa_Si_V, student_Vle)
```

```python
All = pd.merge(C_A_Sa_Si_V_Sv, studentRegistration)
```

```python
All.loc[All['score'] >= 51 , 'final_result'] = 'Pass'
All.loc[All['score'] < 51, 'final_result'] = 'Fail'
```

```python
All.loc[All['score'] < 51, 'ECTS'] = 'F'
All.loc[(All['score'] >= 51) & (All['score'] <= 60), 'ECTS'] = 'E'
All.loc[(All['score'] >= 61) & (All['score'] <= 70), 'ECTS'] = 'D'
All.loc[(All['score'] >= 71) & (All['score'] <= 80), 'ECTS'] = 'C'
All.loc[(All['score'] >= 81) & (All['score'] < 90), 'ECTS'] = 'B'
All.loc[All['score'] >= 90, 'ECTS'] = 'A'
```
